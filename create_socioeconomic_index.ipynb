```python
import pandas as pd
```

```python
econ_df = pd.read_csv('data/socioeconomic_factors.csv', sep=";")
```

```python
new_cols = [col for col in econ_df.columns if ("2021" in col or not "20" in col)]
```

```python
econ_df = econ_df[new_cols]
econ_df.drop("Demografietyp", axis=1, inplace=True)
econ_df
```

```python
import numpy as np
```

```python
for col in econ_df.columns:
    if "2021" not in col:
        continue
    values = econ_df[col]
    s = 0
    c = 0
    for value in values:
        has_kA = False
        if type(value) == str and "k.A." in value:
            has_kA = True
            continue
        if (type(value) == str):
            s += float(value.replace(",", "."))
        else:
            s += value
        c += 1
    mean = s/c
    # print((econ_df[col] == "k.A. ").sum())
    # print(econ_df[col].isna().sum())
    if has_kA:
        econ_df.loc[econ_df[col].str.contains("k.A.", na=False), col] = str(mean)
    print(mean)
```

```python
for col in econ_df.columns:
    if "2021" not in col:
        continue
    try:
        # print(econ_df[col].str.replace(",", "."))
        econ_df[col] = econ_df[col].str.replace(",", ".").astype(float)
    except:
        print("column not string")
```

```python
#econ_df["2021\nAnteil 65- bis 79-J�hrige (%)"] = econ_df["2021\nAnteil 65- bis 79-J�hrige (%)"]
# TODO: Convert to float!!!
# econ_df["2021\nAnteil 65- bis 79-J�hrige (%)"] = econ_df["2021\nAnteil 65- bis 79-J�hrige (%)"].replace(",", ".")
# econ_df["2021\nAnteil ab 80-J�hrige (%)"] = econ_df["2021\nAnteil ab 80-J�hrige (%)"].replace(",", ".")
econ_df["2021\nAnteil ab 65-J�hrige (%)"] = econ_df["2021\nAnteil 65- bis 79-J�hrige (%)"] + econ_df["2021\nAnteil ab 80-J�hrige (%)"]

cols = econ_df.columns

econ_df = econ_df[list(set(econ_df.columns).difference(set(["2021\nAnteil ab 80-J�hrige (%)", "2021\nAnteil 65- bis 79-J�hrige (%)", "2021\nKaufkraft (Euro/Haushalt)"])))]
econ_df
```

```python
# normalize the factors
for col in econ_df.columns:
    if "2021" not in col:
        continue
    econ_df[f"{col}_normalized"] = econ_df[col] / econ_df[col].max()

econ_df["socio_economic_index"] = econ_df["2021\nHaushalte mit niedrigem Einkommen (%)_normalized"] + econ_df["2021\nAnteil ab 65-J�hrige (%)_normalized"] + econ_df["2021\nVerschuldung im Kernhaushalt (Euro je Einwohner:in)_normalized"] + econ_df["2021\nLangzeitarbeitslosenquote (%)_normalized"] - econ_df["2021\nAllgemeine Deckungsmittel (Euro je Einwohner:in)_normalized"]
```
